```python
import csv
import pandas as pd
import requests
```

```python
urldata = 'https://api.tomorrow.io/v4/timelines?location=40.75872069597532,-73.98529171943665&fields=humidity&timesteps=1h&units=metric&apikey=4iVuiItmnsMgySpLVtQYWbCb9eDUHQ57'

header = {'Content-type':'application/json','Accept-Encoding':'deflate'}
repo = requests.get(urldata, headers = header)
print(repo)
```

```python
repoData = repo.json()
print(repoData)
```

**Extract and structure the data**

```python
data_to_write = []
for timeline in repoData['data']['timelines']:
    timestep = timeline['timestep']
    endTime = timeline['endTime']
    startTime = timeline['startTime']
    for interval in timeline['intervals']:
        startTime_interval = interval['startTime']
        temperature = interval['values']['humidity']
        data_to_write.append({
            'timestep': timestep,
            'endTime': endTime,
            'startTime': startTime,
            'startTime_interval': startTime_interval,
            'humidity': temperature
        })

# Define the CSV file path
csv_file_path = 'humidity.csv'

# Write the data to a CSV file
with open(csv_file_path, 'w', newline='') as csv_file:
    fieldnames = ['timestep', 'endTime', 'startTime', 'startTime_interval', 'humidity']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data_to_write)

print(f"Data has been written to {csv_file_path}")
```
