1. In an annual examination result, 10 students’ data (both boys and girls) are taken to study the grades. Using the data, create a Girls vs Boys scatter plot to represent the correlation beween girls and boys. Write a complete python code.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
girls_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
boys_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]

plt.scatter(girls_grades, boys_grades)
plt.xlabel("Girls' Grades")
plt.ylabel("Boys' Grades")
plt.title("Girls vs Boys Grades Scatter Plot")
plt.show()
```

image.png

Write a complete python code that:

a) Creates a line plot using the above Dataframe.


```python

data = {'Name': ['Moses', 'Anne', 'James', 'Ronald', 'Celine'],
        'English': [58, 75, 72, 78, 76],
        'Economics': [71, 81, 85, 85, 84],
        'Maths': [63, 74, 78, 70, 77],
        'Computing': [64, 82, 65, 87,86],
        'Physics': [75, 77, 84, 86, 80]}

DFS = pd.DataFrame(data)
DFS.head()
```

```python
DFS.plot(x='Name', y=['English', 'Economics', 'Maths', 'Computing', 'Physics'], kind='line')
plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Student Marks Line Plot")
plt.show()
```

```python
DFS.boxplot(column=['English', 'Economics', 'Maths', 'Computing', 'Physics'])
plt.title("Student Marks Boxplot")
plt.show()
```


5. A first quarter expense data fir a household is given below:
Expense Category

image.png

a) Write the pyhton code to create a line plot with the following customizations:

i) Set the fogure wodth as 10 and height as 5.

ii) Chart title First Quarter Expenses with fontsize=20.

iii) Xlabel is Expense Category with fontsize=14.

iv) Ylabel is Expenses with fontsize=14.

v) Legend is the three months with fontsize=12.


```python

data = {'Expense Category': ['House Rent', 'Gas', 'Electricity', 'Groceries', 'Insurance', 'Travel', 'Cable', 'Internet', 'Credit Cards'],
        'January': [12500, 910, 1200, 2400, 7000, 2200, 350, 999, 3500],
        'February': [11500, 810, 1350, 2350, 7300, 2350, 350, 999, 3200],
        'March': [11500, 910, 1450, 2700, 7500, 1900, 350, 999, 4500]}
df = pd.DataFrame(data).set_index('Expense Category')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df.index, df['January'], label='January')
ax.plot(df.index, df['February'], label='February')
ax.plot(df.index, df['March'], label='March')
ax.set_title('First Quarter Expenses', fontsize=20)
ax.set_xlabel('Expense Category', fontsize=14)
ax.set_ylabel('Expenses', fontsize=14)
ax.legend(fontsize=12)

plt.show()
```

b) Write the python code to create a bar plot for the data.

```python
fig, ax = plt.subplots(figsize=(10, 5))
df.plot(kind='bar', ax=ax)
ax.set_title('First Quarter Expenses', fontsize=20)
ax.set_xlabel('Expense Category', fontsize=14)
ax.set_ylabel('Expenses', fontsize=14)
ax.legend(fontsize=12)

plt.show()
```
