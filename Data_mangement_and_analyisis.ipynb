**3.4 Handling missing data**


```python
#importig the necesary library
import pandas as pd
#upload the dataset
data = pd.read_csv('Titanic-Dataset.csv')
#display the first five rows
data.head()
```

```python
#check whether there is null values
data.isnull().sum()
```

Handling the missing value by filling using mean manupulation


```python
#Fill in missing values with mean imputation
data['Age'].fillna(data['Age'].mean(), inplace=True)
print(data)
```

**4.1 Crosstabs and pivot tables**

**Crosstabs** Summarizes the relationship between two or more categorical variables also, they are Easy to create and interpret, computationally efficient. While Pivot tables plays avery important role of Summarizing and analyzing data from multiple variables as well as it's Versatile, can handle complex relationships between variables, can calculate complex statistics.

Here is the code of how to implement it.

```python
# Create a crosstab of 'PassengerId' and 'Survived'
crosstab_passenger_id_survived = pd.crosstab(data['PassengerId'], data['Survived'])
print("Crosstab of PassengerId and Survived:")
print(crosstab_passenger_id_survived)

# Create a crosstab of 'Pclass' and 'Embarked'
crosstab_pclass_embarked = pd.crosstab(data['Pclass'], data['Embarked'])
print("Crosstab of Pclass and Embarked:")
print(crosstab_pclass_embarked)

# Create a crosstab of 'Sex' and 'Parch'
crosstab_sex_parch = pd.crosstab(data['Sex'], data['Parch'])
print("Crosstab of Sex and Parch:")
print(crosstab_sex_parch)

# Create a pivot table to summarize the relationship between 'Age', 'Pclass', and 'Fare'
pivot_table_age_pclass_fare = data.pivot_table(index='Age', columns='Pclass', values='Fare', aggfunc='mean')
print("Pivot table summarizing Age, Pclass, and Fare:")
print(pivot_table_age_pclass_fare)

# Create a crosstab of 'Age' and 'Pclass'
crosstab_age_pclass = data.pivot_table(index='Age', columns='Pclass', values='Fare', aggfunc='mean')
# Print the crosstabs
print(crosstab_age_pclass)
```

##**4.6 Introducing regular expressions**



* Cleaning and pre-processing text data: Regular expressions can be used to remove HTML tags, punctuation, and other unwanted characters from text data.

* Extracting information from text data: Regular expressions can be used to extract specific information from text data, such as email addresses, phone numbers, and URLs.

* Validating data: Regular expressions can be used to validate data, such as ensuring that phone numbers have the correct format and that email addresses are well-formed.

#**example of regular expression in python code**

```python
#import regular expresion library
import re

# Define the regular expression pattern
pattern = r"\d{3}-\d{3}-\d{4}"

# Extract phone numbers from text
text = "My phone number is 555-555-5555. Please call me at 123-456-7890."
matches = re.findall(pattern, text)

# Print the extracted phone numbers
print(matches)
```

```python
# Define the regular expression pattern
pattern = r"<.*?>"

# Remove HTML tags from text
text = "<p>This is some text with HTML tags.</p>"
clean_text = re.sub(pattern, "", text)

# Print the clean text
print(clean_text)
```

##**5.1 Anscombeâ€™s Quartet - visualising data**

```python
#import library
import matplotlib.pyplot as plt
import numpy as np
# Load the data
data = pd.read_csv("anscombe.csv")

# Calculate the number of rows needed for subplots
num_rows = (data.shape[0] + 1) // 2
# Create histograms
fig, axs = plt.subplots(num_rows, 2, figsize=(10, 10))

# Iterate through rows of data and plot on existing subplots
for i, row in data.iterrows():
    row_index = i // 2
    col_index = i % 2
    if row_index < num_rows:
        axs[row_index, col_index].hist(row[1:])

# Label axes and show plot
plt.xlabel("X")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

##Let us go a head to visualize the anscombe data using boxplot

```python
# Calculate the number of rows needed for subplots
num_rows = (data.shape[0] + 1) // 2

# Create box plots
fig, axs = plt.subplots(num_rows, 2, figsize=(10, 10))

# Iterate through rows of data and plot box plots on existing subplots
for i, row in data.iterrows():
    row_index = i // 2
    col_index = i % 2
    if row_index < num_rows:
          #plot the boxplot
        axs[row_index, col_index].boxplot(row[1:], vert=False)

# Label axes and show plot
plt.xlabel("X")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```

##Scatter plots for the data.


```python
# Calculate the number of rows needed for subplots
num_rows = (data.shape[0] + 1) // 2

# Create scatter plots
fig, axs = plt.subplots(num_rows, 2, figsize=(10, 10))
# Iterate through rows of data and plot scatter plots on existing subplots
for i, row in data.iterrows():
    row_index = i // 2
    col_index = i % 2
    if row_index < num_rows:
        axs[row_index, col_index].scatter(range(1, len(row)), row[1:])

# Label axes and show plot
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

## This line plot shows clearly how the data is behaving and its relationship.


```python
# Create scatter plots with line graphs
fig, axs = plt.subplots(num_rows, 2, figsize=(10, 10))

# Iterate through rows of data and plot scatter plots and line graphs on existing subplots
for i, row in data.iterrows():
    row_index = i // 2
    col_index = i % 2
    if row_index < num_rows:
        x_values = range(1, len(row))
        axs[row_index, col_index].scatter(x_values, row[1:], alpha=0.5, label='Scatter')
        axs[row_index, col_index].plot(x_values, row[1:], color='red', label='Line')

# Label axes and show plot
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

##**5.2 Getting started with maps â€“ folium**



**Installation of the folium library**

```python
import folium
from folium.plugins import MarkerCluster
```

In this step we create a sample map by  puting the coordinate of the city in Saudi Arabia.

```python
# Create a map centered at a specific location
latitude = 	24.6667
longitude = 46.6958
simple_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Display the map
simple_map
```

**Let now customize the map**

```python
# Customize the map by changing the tileset
customized_map = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='Stamen Terrain')

# Display the customized map
customized_map
```

**Add makers to the map**

This step is associated with a location (latitude, longitude) and a popup message.

```python
# Create a map with markers
marker_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Add markers to the map
folium.Marker([latitude + 0.01, longitude], popup='Marker 1').add_to(marker_map)
folium.Marker([latitude, longitude + 0.01], popup='Marker 2').add_to(marker_map)

# Display the map with markers
marker_map
```

**Creating a Heatmap**

This code is capable of generating random data and creates a heatmap layer on the map using the HeatMap plugin from Folium libraly.

```python
# Generate random data for a heatmap
heat_data = [[latitude + 0.001, longitude + 0.001, 1],
             [latitude - 0.002, longitude - 0.002, 2],
             [latitude + 0.003, longitude - 0.001, 3]]

# Create a map with a heatmap layer
heatmap_map = folium.Map(location=[latitude, longitude], zoom_start=12)
folium.plugins.HeatMap(heat_data).add_to(heatmap_map)

# Display the map with the heatmap layer
heatmap_map
```

**Next step is Combine Marker and Heatmap**

 Here, we combine both markers and a heatmap layer on the same map.

```python
# Create a map with both markers and a heatmap
combined_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Add markers to the map
folium.Marker([latitude + 0.01, longitude], popup='Marker 1').add_to(combined_map)
folium.Marker([latitude, longitude + 0.01], popup='Marker 2').add_to(combined_map)

# Add a heatmap layer to the map
folium.plugins.HeatMap(heat_data).add_to(combined_map)

# Display the map with markers and heatmap
combined_map
```

```python
#Loading the dataset
movies_df = pd.read_csv('imdb_movies.csv')

# Display the first few rows of the dataset
movies_df.head()
```

**Data exploration**

```python
#checking the statistics in the dataset
summary_stats = movies_df.describe()

# Get information about the dataset, such as data types and missing values
data_info = movies_df.info()

# Explore unique values in a specific column let say the 'genre'
unique_genres = movies_df['genre'].unique()
```

**Data visualization**

This focuses on the genre column and the number of movies

```python
# Plot a bar chart of movie genres
plt.figure(figsize=(10, 8))
sns.countplot(x='genre', data=movies_df, order=movies_df['genre'].value_counts().index)
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

As you can observe the above graph the genre movies compared with the number of movies it seems to be higher.

**Analyzing the relationship**

```python
# Create a scatter plot to analyze the relationship between 'budget_x' and 'revenue'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget_x', y='revenue', data=movies_df)
plt.title('Relationship between Budget and Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

**Conduct Analysis**

```python
# Calculate the average revenue for each genre
avg_revenue_by_genre = movies_df.groupby('genre')['revenue'].mean().sort_values(ascending=False)

# Display the result
print("Average Revenue by Genre:\n", avg_revenue_by_genre)
```

From the above results you can clearly see that the revenue  for the Adventure, Action, Science Fiction, Fantasy is extreamly high of about 2.068224e+09 when compare to the other genre.
