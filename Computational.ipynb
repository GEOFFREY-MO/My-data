Let now import numpy libray and declare all patient variables

then disply the results interms of the mean and compare it with our excel sheet above.

```python
import numpy as np
A,B,C,D,E=np.array([[40,37,39,37,30],[39,35,37,40,37.5],[37,36,40,39,41],[36,38,39,38,39],[38,39,37,39,39]])
print(A,B,C,D,E)
```

```python
Q = A,B,C,D,E
```

```python
Q
```

```python
mean=np.mean(Q,axis=1)
```

```python
mean
```

As you can see the results of the mean or average of the code above is exactly

as it is in the yellow column of excel sheet.

Lets now apply **where()** condition since we are dealing with an array with mulptiple  columns.

```python
patients = mean[np.where(np.logical_and(mean>37,mean<50))]
patients2 = mean[np.where(np.logical_and(mean>10,mean<37))]
print('Discharge',patients2)
print('Admit',patients)
```
